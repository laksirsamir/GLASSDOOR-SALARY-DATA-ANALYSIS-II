```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./salary_data_cleaned/salary_data_cleaned.csv')

# Group by company and calculate average salary
salary_by_company = data.groupby('company_txt')['avg_salary'].mean().reset_index()

# Sort the companies by average salary in descending order and select the top 10
top_companies = salary_by_company.sort_values(by='avg_salary', ascending=False).head(10)

# Display the top companies with the highest average salaries
print(top_companies)

# Plot the top companies with the highest average salaries
plt.figure(figsize=(14, 8))
sns.barplot(x='avg_salary', y='company_txt', data=top_companies, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Company')
plt.title('Top Companies with Highest Average Salaries')
plt.show()
```
