```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./salary_data_cleaned/salary_data_cleaned.csv')

# Group by sector and calculate average salary
salary_stats_by_sector = data.groupby('Sector')['avg_salary'].mean().reset_index()

# Identify the top sector by average salary
top_sector = salary_stats_by_sector.loc[salary_stats_by_sector['avg_salary'].idxmax()]
print(f"Top Sector: {top_sector['Sector']} with Average Salary: {top_sector['avg_salary']}")

# Filter the data for the top sector
top_sector_data = data[data['Sector'] == top_sector['Sector']]

# Calculate salary statistics within the top sector
salary_stats_within_top_sector = top_sector_data['avg_salary'].describe()
print(salary_stats_within_top_sector)

# Plot the salary distribution within the top sector using a histogram
plt.figure(figsize=(12, 8))
sns.histplot(top_sector_data['avg_salary'], bins=20, kde=True, color='blue')
plt.xlabel('Average Salary')
plt.ylabel('Frequency')
plt.title(f'Salary Distribution in the {top_sector["Sector"]} Sector')
plt.show()

# Plot the salary distribution within the top sector using a box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='avg_salary', data=top_sector_data, color='blue')
plt.xlabel('Average Salary')
plt.title(f'Salary Distribution in the {top_sector["Sector"]} Sector')
plt.show()
```

```python
# Display unique sectors in the dataset
unique_sectors = data['Sector'].unique()
print("Unique Sectors in the Dataset:")
for sector in unique_sectors:
    print(sector)
```

```python
# Specify the sectors to analyze
chosen_sectors = ['Information Technology', 'Telecommunications', 'Aerospace & Defense', 'Insurance']  # Example sectors

# Filter the data for the specified sectors
filtered_data = data[data['Sector'].isin(chosen_sectors)]

# Display the first few rows of the filtered dataset
print(filtered_data.head())

# Calculate salary statistics within the specified sectors
salary_stats_within_sectors = filtered_data.groupby('Sector')['avg_salary'].describe()
print(salary_stats_within_sectors)

# Plot the salary distribution within the specified sectors using a histogram
plt.figure(figsize=(12, 8))
sns.histplot(filtered_data, x='avg_salary', hue='Sector', bins=20, kde=True, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Frequency')
plt.title(f'Salary Distribution in the Selected Sectors')
plt.show()

# Plot the salary distribution within the specified sectors using a box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='avg_salary', y='Sector', data=filtered_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Sector')
plt.title(f'Salary Distribution in the Selected Sectors')
plt.show()
```
