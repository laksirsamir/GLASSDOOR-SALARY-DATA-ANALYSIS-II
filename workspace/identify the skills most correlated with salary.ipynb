```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./salary_data_cleaned/salary_data_cleaned.csv')

# Extract relevant skill columns and the salary column
skills = ['python_yn', 'R_yn', 'spark', 'aws', 'excel']
salary = 'avg_salary'

# Calculate correlations between skills and salary
correlations = data[skills + [salary]].corr()

# Extract the correlations with the salary
skill_correlations = correlations[salary].drop(salary)
print(skill_correlations)

# Plot the correlations
plt.figure(figsize=(10, 6))
sns.barplot(x=skill_correlations.values, y=skill_correlations.index, palette='viridis')
plt.xlabel('Correlation with Average Salary')
plt.ylabel('Skill')
plt.title('Skills Most Correlated with Salary')
plt.show()
```
