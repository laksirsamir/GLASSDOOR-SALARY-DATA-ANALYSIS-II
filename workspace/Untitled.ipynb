```python
import pandas as pd

# Load the dataset
data = pd.read_csv('./salary_data_cleaned/salary_data_cleaned.csv')

# Display the first few rows of the dataset and its summary
print(data.head())
print(data.info())
print(data.describe())
print(data.columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by industry and calculate salary statistics
salary_stats_by_industry = data.groupby('Industry')['avg_salary'].describe()
print(salary_stats_by_industry)

# Plot the salary distribution across different industries using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='avg_salary', y='Industry', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.title('Salary Distribution Across Different Industries')
plt.show()

# Plot the salary distribution across different industries using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='avg_salary', y='Industry', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.title('Salary Distribution Across Different Industries')
plt.show()

# Group by job title and calculate salary statistics
salary_stats_by_job_title = data.groupby('Job Title')['avg_salary'].describe()
print(salary_stats_by_job_title)

# Plot the salary distribution across different job titles using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='avg_salary', y='Job Title', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Salary Distribution Across Different Job Titles')
plt.show()

# Plot the salary distribution across different job titles using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='avg_salary', y='Job Title', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Salary Distribution Across Different Job Titles')
plt.show()

# Group by company size and calculate salary statistics
salary_stats_by_size = data.groupby('Size')['avg_salary'].describe()
print(salary_stats_by_size)

# Plot the salary distribution across different company sizes using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='avg_salary', y='Size', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Company Size')
plt.title('Salary Distribution Across Different Company Sizes')
plt.show()

# Plot the salary distribution across different company sizes using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='avg_salary', y='Size', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Company Size')
plt.title('Salary Distribution Across Different Company Sizes')
plt.show()

# Group by location and calculate salary statistics
salary_stats_by_location = data.groupby('Location')['avg_salary'].describe()
print(salary_stats_by_location)

# Plot the salary distribution across different locations using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='avg_salary', y='Location', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Location')
plt.title('Salary Distribution Across Different Locations')
plt.show()

# Plot the salary distribution across different locations using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='avg_salary', y='Location', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Location')
plt.title('Salary Distribution Across Different Locations')
plt.show()
```
