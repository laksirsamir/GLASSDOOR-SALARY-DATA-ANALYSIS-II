```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./salary_data_cleaned/salary_data_cleaned.csv')

# Group by sector and calculate average salary
salary_stats_by_sector = data.groupby('Sector')['avg_salary'].describe()
print(salary_stats_by_sector)

# Plot the salary distribution across different sectors using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='avg_salary', y='Sector', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Sector')
plt.title('Salary Distribution Across Different Sectors')
plt.show()

# Plot the salary distribution across different sectors using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='avg_salary', y='Sector', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Sector')
plt.title('Salary Distribution Across Different Sectors')
plt.show()
```

```python
# Filter by specific sectors
sectors_of_interest = [
    'Business Services', 
    'Education', 
    'Finance',
    'Government',
    'Health Care',
    'Information Technology',
    'Telecommunications'
]

filtered_data = data[data['Sector'].isin(sectors_of_interest)]

# Display the first few rows of the filtered dataset
print(filtered_data.head())

# Group by sector and calculate average salary
salary_stats_by_sector = filtered_data.groupby('Sector')['avg_salary'].describe()
print(salary_stats_by_sector)

# Plot the salary distribution across different sectors using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='avg_salary', y='Sector', data=filtered_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Sector')
plt.title('Salary Distribution Across Different Sectors')
plt.show()

# Plot the salary distribution across different sectors using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='avg_salary', y='Sector', data=filtered_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Sector')
plt.title('Salary Distribution Across Different Sectors')
plt.show()
```
