```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./salary_data_cleaned/salary_data_cleaned.csv')

# Display the first few rows of the dataset and its summary
print(data.head())
print(data.columns)

# Group by job title and calculate average salary
salary_stats_by_job_title = data.groupby('Job Title')['avg_salary'].describe()
print(salary_stats_by_job_title)

# Plot the salary distribution across different job titles using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='avg_salary', y='Job Title', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Salary Distribution Across Different Job Titles')
plt.show()

# Plot the salary distribution across different job titles using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='avg_salary', y='Job Title', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Salary Distribution Across Different Job Titles')
plt.show()
```

```python

# Filter by specific job titles
job_titles_of_interest = [
    'Ag Data Scientist', 
    'Analytics Consultant', 
    'Analytics Manager',
    'System and Data Analyst',
    'Web Data Analyst'
]

filtered_data = data[data['Job Title'].isin(job_titles_of_interest)]

# Display the first few rows of the filtered dataset
print(filtered_data.head())

# Group by job title and calculate average salary
salary_stats_by_job_title = filtered_data.groupby('Job Title')['avg_salary'].describe()
print(salary_stats_by_job_title)
```

```python
# Plot the salary distribution across different job titles using bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='avg_salary', y='Job Title', data=filtered_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Average Salary by Job Title')
plt.show()

# Plot the salary distribution across different job titles using swarm plot
plt.figure(figsize=(12, 8))
sns.swarmplot(x='avg_salary', y='Job Title', data=filtered_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Salary Distribution by Job Title')
plt.show()

# Plot the salary distribution across different job titles using point plot
plt.figure(figsize=(12, 8))
sns.pointplot(x='avg_salary', y='Job Title', data=filtered_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Average Salary by Job Title with Confidence Intervals')
plt.show()
```
