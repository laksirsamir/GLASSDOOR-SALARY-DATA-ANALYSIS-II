```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('./salary_data_cleaned/salary_data_cleaned.csv')

# Ensure 'Founded' column contains only valid years
data = data[data['Founded'] > 0]

# Group by year and calculate average salary
salary_trends_by_year = data.groupby('Founded')['avg_salary'].mean().reset_index()

# Display the salary trends by year
print(salary_trends_by_year)

# Plot the salary trends over years
plt.figure(figsize=(14, 8))
plt.plot(salary_trends_by_year['Founded'], salary_trends_by_year['avg_salary'], marker='o')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.title('Salary Trends Over Years')
plt.grid(True)
plt.show()
```
